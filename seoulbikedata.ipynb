```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif(df[[i for i in df.describe().columns if i not in ['Rented Bike Count','Dew point temperature(°C)']]])
```

```python
df1 = pd.get_dummies(df,columns = ['Seasons','Holiday', 'Functioning Day', 'year', 'month', 'day','Hour'],drop_first=True)
```

```python
df1.drop('Dew point temperature(°C)',axis=1,inplace=True)
```

```python
df1.head()
```

```python
X = df1.drop('Rented Bike Count',axis = 1)
y = df['Rented Bike Count']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 10)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print('Training r2_score')
print(lr.score(X_train,y_train))
print('Testing r2_score')
print(lr.score(X_test,y_test))
```

```python
np.mean(cross_val_score(lr,X_train,y_train,cv=10))
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer()
X_train_transformed = pt.fit_transform(X_train)
X_test_transformed = pt.transform(X_test)
lr1 = LinearRegression()
lr1.fit(X_train_transformed,y_train)
y_pred1 = lr1.predict(X_test_transformed)
r2_score(y_test,y_pred1)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge = Ridge()
ridge_regressor = GridSearchCV(ridge,parameters,scoring = 'neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
ridge_regressor.best_params_
```

```python
y_pred2  =ridge_regressor.predict(X_test)
```

```python
r2_score(y_pred2,y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline((
    ('pt',PowerTransformer()),('lr',LinearRegression())
))
pipe.fit(X_train,y_train)
pipe.score(X_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
pipe1 = Pipeline((
    ('pt',PowerTransformer()),('gbr',GradientBoostingRegressor())
))
pipe1.fit(X_train,y_train)
pipe1.score(X_train,y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe2 = Pipeline((
    ('poly',PolynomialFeatures(degree=2)),('lr',LinearRegression())
))
pipe2.fit(X_train,y_train)
pipe2.score(X_train,y_train)
pipe2.score(X_test,y_test)
```

```python
pipe3 = Pipeline((
    ('poly',PolynomialFeatures(degree=3)),('lr',LinearRegression())
))
pipe3.fit(X_train,y_train)
pipe3.score(X_train,y_train)
```

```python
pipe3.score(X_test,y_test)
```
